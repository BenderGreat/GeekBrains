#### 1) Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].  Найдите ее среднее значение и дисперсию.

Математическое ожидание $M(X)$ равномерно распределенной непрерывной случайной величины можно вычислить по формуле:

$$M(X) = \frac{a + b}{2}$$

По следующей формуле можно рассчитать дисперсию $D(X)$:

$$D(X) = \frac{(b - a)^2}{12}$$

```python
a = 200
b = 800
print(f'Cреднее значение cлучайной непрерывной величины A  = {(a+b)/2}')
```

```python
print(f'Дисперсия cлучайной непрерывной величины A  = {(b-a)**2/12}')
```

#### 3) Непрерывная случайная величина X распределена нормально и задана плотностью распределения
#### f(x) = (1 / (4 * sqrt(2*pi))) * (exp(-((x+2)**2)/ 32)).
#### Найдите:
#### а). M(X)
#### б). D(X)
#### в). std(X) (среднее квадратичное отклонение)

Нормальное распределение — это распределение вероятностей непрерывной случайной величины $X$, плотность вероятности которой подчиняется формуле:

$$f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e ^ {- \frac{(x - a)^2}{2 \sigma^2}},$$

где $a = M(X), \; \sigma^2 = D(X).$

```python
sigma = 4
MX = -2
DX = 16
print(f'M(X) = {MX}')
```

```python
print(f'D(X) = {DX}')
```

```python
print(f'std(X) (среднее квадратичное отклонение) = {sigma}')
```

```python
X = 190
MX = 178
DX = 25
sigma = sqrt(DX)
print(f'Среднее квадратичное отклонение = {sigma}')
```

```python
print(f'Рост человека равный 190 см отклоняется от мат ожидания роста в популяции на {abs((MX - X)/sigma)} сигм')
```
